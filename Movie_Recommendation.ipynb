```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
movies.head(1)
```

```python
movies=movies.merge(credits,on='title')
movies.head()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew',
                 'popularity', 'spoken_languages', 'vote_average', 'runtime']]
```

```python
print(movies['spoken_languages'][0])
```

```python
import ast

def extract_names(text):
    try:
        return [i['name'] for i in ast.literal_eval(text)]  # ✅ fixed 'names' → 'name'
    except:
        return []

def extract_top_cast(text):
    try:
        return [i['name'] for i in ast.literal_eval(text)[:3]]  # ✅ fixed function name
    except:
        return []

def extract_director(text):
    try:
        for i in ast.literal_eval(text):
            if i['job'] == 'Director':
                return [i['name']]
        return []
    except:
        return []
```

```python
movies['genres'] = movies['genres'].apply(extract_names)
movies['keywords'] = movies['keywords'].apply(extract_names)
movies['cast'] = movies['cast'].apply(extract_top_cast)
movies['crew'] = movies['crew'].apply(extract_director)
movies['spoken_languages'] = movies['spoken_languages'].apply(extract_names)
movies.head()
```

```python
def collapse(lst):
  return " ".join(lst)

movies['overview']=movies['overview'].fillna("")
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['tags'] = movies['overview'] + " " + movies['genres'] + " " + movies['keywords'] + " " + movies['cast'] + " " + movies['crew']
print(movies['tags'][0])
```

```python
cv = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
print(vectors.shape)
print(vectors[0])
```

```python
similarity = cosine_similarity(vectors)
print(similarity.shape)
print(similarity[0])
```

```python
def recommend(movie):
    movie = movie.lower()
    matches = movies[movies['title'].str.lower() == movie]

    if matches.empty:
        print("Sorry, movie not found 😢")
        return

    movie_index = matches.index[0]
    distances = similarity[movie_index]

    movie_list = list(enumerate(distances))
    movie_list = sorted(movie_list, key=lambda x: x[1], reverse=True)[1:6]

    for i in movie_list:
        print(movies.iloc[i[0]].title)

```

```python
recommend("Fight Club")
```

```python
movies.to_csv('movies.csv', index=False)
```
